```python
import pandas as pd
import os
import subprocess

# 读取你的歌曲列表
df = pd.read_csv("/Users/xiahanfei/Desktop/MGS3001-Python-for-Business/AI_music/netease_songs_final_complete.csv")

# 创建保存文件夹
os.makedirs("song_audio", exist_ok=True)

# 开始全自动下载
for index, row in df.iterrows():
    song = row["song_name"]
    artist = row["artist_name"]
    
    # 搜索关键词：歌名 + 歌手
    keyword = f"{song} {artist} audio"
    
    # 保存文件名：歌手 - 歌名
    filename = f"{artist} - {song}"
    filepath = f"song_audio/{filename}.mp3"

    # 如果已经下载过，跳过
    if os.path.exists(filepath):
        print(f"✅ 已存在：{filename}")
        continue

    print(f"⏬ 正在下载：{filename}")

    # 自动从YouTube下载高清音频（最稳定、无版权风险、学术可用）
    command = [
        "yt-dlp",
        "-x", "--audio-format", "mp3",
        "--audio-quality", "192K",
        "-o", f"song_audio/{filename}.%(ext)s",
        f"ytsearch1:{keyword}"
    ]

    try:
        subprocess.run(command, check=True, capture_output=True)
        print(f"✅ 下载完成：{filename}\n")
    except:
        print(f"❌ 下载失败：{filename}（换个平台即可）\n")
```
