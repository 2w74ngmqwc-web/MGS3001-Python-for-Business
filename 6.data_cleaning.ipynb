```python
import pandas as pd

# ======================
# 路径配置
# ======================
INPUT_FILE  = "/Users/xiahanfei/Desktop/MGS3001-Python-for-Business/AI_music/AIGC rate/Audio_AIGC rate_FINAL.csv"
OUTPUT_FILE = "/Users/xiahanfei/Desktop/MGS3001-Python-for-Business/AI_music/AIGC rate/Audio_AIGC rate_FINAL_cleaned.csv"

# ======================
# 1. 读取原始数据
# ======================
df = pd.read_csv(INPUT_FILE, encoding="utf-8-sig")
print("📊 清洗前数据概况：")
print(f"总样本数：{len(df)} 首")
print(f"音频AI率为0的样本数：{len(df[df['Audio_AIGC_Rate'] == 0])} 首")
print(f"歌词AI率为0的样本数：{len(df[df['Lyrics_AIGC_Rate'] == 0])} 首")

# ======================
# 2. 数据清洗（核心筛选）
# ======================
# 只保留：音频AI率>0 且 歌词AI率>0 的有效样本
df_cleaned = df[
    (df["Audio_AIGC_Rate"] > 0) & 
    (df["Lyrics_AIGC_Rate"] > 0)
].copy()

# 重置索引（避免筛选后索引断层）
df_cleaned = df_cleaned.reset_index(drop=True)

# ======================
# 3. 清洗后数据统计
# ======================
print("\n✅ 清洗后数据概况：")
print(f"有效样本数：{len(df_cleaned)} 首")
print(f"筛选掉无效样本数：{len(df) - len(df_cleaned)} 首")
print(f"\n核心变量描述统计：")
print(df_cleaned[["Audio_AIGC_Rate", "Lyrics_AIGC_Rate", "Total_AI_Permeability"]].describe().round(3))

# ======================
# 4. 保存清洗后的文件
# ======================
df_cleaned.to_csv(OUTPUT_FILE, index=False, encoding="utf-8-sig")
print(f"\n💾 清洗后文件已保存：")
print(OUTPUT_FILE)

# 预览前5行有效数据
print("\n🔍 清洗后数据预览：")
print(df_cleaned[["artist_name", "song_name", "Audio_AIGC_Rate", "Lyrics_AIGC_Rate", "Total_AI_Permeability"]].head())
```

```python
import pandas as pd

# ======================
# 路径配置（用你最终的干净数据文件）
# ======================
INPUT_FILE  = "/Users/xiahanfei/Desktop/MGS3001-Python-for-Business/AI_music/creativity/Usefulness_Novelty.csv"
OUTPUT_FILE = "/Users/xiahanfei/Desktop/MGS3001-Python-for-Business/AI_music/creativity/Usefulness_Novelty_final_no_novelty1.csv"

# ======================
# 1. 读取数据 + 统计初始情况
# ======================
df = pd.read_csv(INPUT_FILE, encoding="utf-8-sig")
original_count = len(df)
# 统计 Novelty=1.0 的样本数
novelty1_count = len(df[df["Novelty"] == 1.0])

print("📊 清洗前概况：")
print(f"总样本数：{original_count} 行")
print(f"Novelty=1.0 的样本数：{novelty1_count} 行")
print(f"Novelty=1.0 占比：{round(novelty1_count/original_count*100, 2)}%")

# ======================
# 2. 核心清洗：移除 Novelty=1.0 的行
# ======================
df_cleaned = df[df["Novelty"] != 1.0].copy()
# 重置索引（避免筛选后索引断层）
df_cleaned = df_cleaned.reset_index(drop=True)

# ======================
# 3. 清洗后结果统计
# ======================
final_count = len(df_cleaned)
print(f"\n✅ 清洗后概况：")
print(f"剩余样本数：{final_count} 行")
print(f"成功删除：{original_count - final_count} 行（全部为 Novelty=1.0 的样本）")
print(f"剩余样本中 Novelty=1.0 的数量：{len(df_cleaned[df_cleaned['Novelty'] == 1.0])} 行（已清零）")

# 展示清洗后 Novelty 列的分布（验证效果）
print(f"\n🔍 清洗后 Novelty 列分布（前5个常见值）：")
novelty_dist = df_cleaned["Novelty"].value_counts().head(5)
print(novelty_dist)

# ======================
# 4. 保存最终文件
# ======================
df_cleaned.to_csv(OUTPUT_FILE, index=False, encoding="utf-8-sig")
print(f"\n💾 最终文件已保存：{OUTPUT_FILE}")
print("\n📌 数据预览（前5行，Novelty 均≠1.0）：")
print(df_cleaned[["artist_name", "song_name", "Usefulness", "Novelty"]].head())
```

```python
import pandas as pd
import os

# 修正后的文件路径（用你给的正确文件名）
ai_rate_path = "/Users/xiahanfei/Desktop/MGS3001-Python-for-Business/AI_music/AIGC rate/Audio_AIGC rate_FINAL.csv"
score_path = "/Users/xiahanfei/Desktop/MGS3001-Python-for-Business/AI_music/creativity/Usefulness_Novelty_final_no_novelty1.csv"

print("检查文件是否存在：")
print(f"AI率文件：{os.path.exists(ai_rate_path)}")
print(f"评分文件：{os.path.exists(score_path)}")

if os.path.exists(ai_rate_path) and os.path.exists(score_path):
    # 读取数据
    df_ai = pd.read_csv(ai_rate_path, encoding="utf-8-sig")
    df_score = pd.read_csv(score_path, encoding="utf-8-sig")
    print("✅ 两个文件读取成功！")

    # 合并数据
    df_final_analysis = pd.merge(
        df_ai,
        df_score,
        on=["artist_name", "song_name"],
        how="inner"
    )

    # 保存文件
    output_path = "/Users/xiahanfei/Desktop/MGS3001-Python-for-Business/AI_music/FINAL_ANALYSIS_DATA.csv"
    df_final_analysis.to_csv(output_path, index=False, encoding="utf-8-sig")

    # 打印结果
    print(f"合并前AI率数据量：{len(df_ai)}")
    print(f"合并前评分数据量：{len(df_score)}")
    print(f"合并后数据量：{len(df_final_analysis)}")
    print(f"✅ 合并完成，文件已保存到：{output_path}")
else:
    print("❌ 文件路径还是不对！请再核对一下文件名。")
```
